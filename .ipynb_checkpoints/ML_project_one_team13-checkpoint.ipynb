```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Read the original data files
jeopardy_adult_df= pd.read_csv('master_season1-35.tsv', sep='\t')
jeopardy_adult_df
```

```python
# Read the data files, change air_date to date to use for sorting & making it the index
jeopardy_df= pd.read_csv('master_season1-35.tsv', sep='\t', infer_datetime_format=True, parse_dates=True, index_col="air_date")
jeopardy_df
```

```python
#delete un-needed columns
jeopardy_df.drop('comments', axis=1, inplace=True)
jeopardy_df.drop('notes', axis=1, inplace=True)
jeopardy_df.drop('value', axis=1, inplace=True)
jeopardy_df.drop('daily_double', axis=1, inplace=True)
jeopardy_df.drop('answer', axis=1, inplace=True)
jeopardy_df.drop('question', axis=1, inplace=True)
jeopardy_df
#df.drop('column_name', axis=1, inplace=True)
```

```python
finals_df = jeopardy_df.loc[jeopardy_df["round"] == 3]
#finals_df.to_csv("../output/finals.csv", index=True)
finals_df
```

```python
early_df = jeopardy_df.loc[(jeopardy_df["round"] == 1) | (jeopardy_df["round"] == 2)]
#finals_df.to_csv("../output/finals.csv", index=True)
early_df
```

```python
#pull out category values from air dates in 1984-1989
early_80 = early_df['1984':'1989'].groupby('category').count().sort_values('round', ascending=False)
final_80 = finals_df['1984':'1989'].groupby('category').count().sort_values('round', ascending=False)
print(early_80)
print(final_80)           #**somehow take only top5 & problems with index

categ_80_df =pd.DataFrame ([{"early_1980s":early_80, "finals_1980s":final_80}], index=0)   #columns=list('t')
categ_80_df
#categ_80.plot(kind='bar')
#categ_early_df

#state_summary_df = pd.DataFrame({"Number of Sightings": state_counts, "Total Visit Time": state_duration})
#df = pd.DataFrame.from_records([{ 'A':a,'B':b }], index='A')   by=['name','timestamp']
```
